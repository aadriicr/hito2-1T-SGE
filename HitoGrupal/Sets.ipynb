## Sets

```python
#Creacion de un Set
set = {"madrid",True,18,9.54,'sevilla'}
print(set)
```

### Metodos

```python
#add(): a√±ade un elemento
set.add(85)
print(set)
```

```python
#clear(): elimina todos los elementos
set.clear()
print(set)
```

```python
#copy(): copia un set
set = {"madrid",True,18,9.54,'sevilla'}
set1 = set.copy()
set == set1
print(set)
print(set1)
```

```python
#difference(): te muestra los elementos distintos entre dos sets
set.add("Barcelona")
print(set.difference(set1))
```

```python
#difference_update(): actualiza el set con los elementos diferentes de otro set
set.add("Valencia")
set1.difference_update(set)
print(set)
```

```python
#discard(): elimina un elemento se la lista
set.discard('madrid')
print(set)
```

```python
#intersection(): devuelve la interseccion entre varios conjuntos mutables
print(set)
print (set1)
print (set.intersection(set1))
```

```python
#pop(): elimina el ultimo elemento de la lista o el elemento seleccionado
print(set.pop())
print(set)
```

```python
#remove(): sirve para borrar un elemento concreto en el caso de tener varios elementos iguales borra el primero
set.remove("Barcelona")
print(set)
```

```python
#update(): sirve para borrar un elemento concreto en el caso de tener varios elementos iguales borra el primero
conjunto={1,2}
set.update(conjunto)
print(set)
```
