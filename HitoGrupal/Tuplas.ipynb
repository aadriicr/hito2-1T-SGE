## Tuplas

```python
#Creacion de la Tupla
tupla=("madrid",True,18,9.54,'sevilla')
#Pintamos la Tupla
print(tupla)
```

### Metodos

```python
#count(): cuenta el numero de veces que aparece un elemento
tupla.count(True)
```

```python
#index(): te indica la posicion del primer elemento seleccionado
print(tupla.index(True))
```
