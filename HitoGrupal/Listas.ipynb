## Listas

```python
#Creacion de la Lista
lista=["madrid",True,18,9.54,'sevilla']
#Pintamos la Lista
print(lista)
#Pintamos el segundo elemento de la lista
print(lista[1])
#Numero de elementos de la Lista
print(len(lista))
#Añadir elemento
lista[2]="Barcelona"
print(lista)
```

### Metodos

```python
#append(): añade un elemento en la ultima posición
lista.append("madrid")
print(lista)
```

```python
#count(): cuenta el numero de veces que aparece un elemento
lista.count("madrid")
```

```python
#extend(): añade un elemento iterable en la ultima posición
lista.extend([3])
print(lista)
```

```python
#index(): te indica la posicion del primer elemento seleccionado 
lista.index("madrid")
```

```python
#insert(): inserta in elemento en la lista podiendo elegir la posicion y el elemento que vamos a añadir
lista.insert(2,"Valencia")
print(lista)
```

```python
#pop(): elimina el ultimo elemento de la lista o el elemento seleccionado
print(lista.pop())
print(lista.pop(2))
print(lista)
```

```python
#remove(): sirve para borrar un elemento concreto en el caso de tener varios elementos iguales borra el primero
lista.remove("Barcelona")
print(lista)
```

```python
#reverse(): invierte el orden de los elementos de la lista
lista.reverse()
print(lista)
```

```python
#sort(): ordena los elementos de una lista (no ordena datos tipos float y string)
numeros=[6,3.8,2,4.7,5,6.4]
numeros.sort()
print(numeros)
```
