## Diccionario

```python
#Creacion del diccionario
diccionario={
    "marca":"bmw",
    "modelo":"x3",
    "diesel":True,
    "consumo":6.9
}
#Pintamos el diccionario
print (diccionario)
#Pintamos la marca
print(diccionario["marca"])
#Numero de elementos de la Lista
print(len(diccionario))
```

### Metodos

```python
#clear(): remueve todos los elementos desde el diccionario.
diccionario.clear()
print(diccionario)
```

```python
#copy(): devuelve una copia superficial del tipo diccionario.
diccionario1= diccionario.copy()
diccionario==diccionario1
```

```python
#get(): devuelve el valor en base a una en un diccionario mediante una clave
diccionario.get("modelo")
```

```python
#items() : devuelve una lista de pares de diccionarios como 2 tuplas
diccionario2= dict(año=2002, kilometros=196000, aceleracion=10.1)
diccionario2.items()
```

```python
#keys() : devuelve una lista de las claves del diccionario.
diccionario2= dict(año=2002, kilometros=196000, aceleracion=10.1)
diccionario2.keys()
```

```python
#pop() : remueve una clave de diccionario y devuelve el valor correspondiente
diccionario2.pop("kilometros")
print(diccionario2)
```

```python
#update() :actualiza el diccionario agregando clave-valores en otro diccionario
diccionario_adicional=dict(aceleracion=9.1)
#print(diccionario_adicional)
diccionario2.update(diccionario_adicional)
print(diccionario2)
```

```python
#values() : devuelve una lista de los valores del diccionario
diccionario2.values()
```
